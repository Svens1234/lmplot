```python
import seaborn as sns
tips = sns.load_dataset("tips")
g = sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
               markers=["o", "x"])
```

```python
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
               palette="Set1")
```
